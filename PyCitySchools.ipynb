```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total number of schools
total_school = len(school_data_complete["school_name"].unique())

# Total number of students
total_students = school_data_complete["Student ID"].count()

# Total budget
total_budget = school_data["budget"].sum()

# Average math score
avg_math = school_data_complete["math_score"].mean()

# Average reading score
avg_reading = school_data_complete["reading_score"].mean()

# Percentage of students with a passing math score (70 or greater)
students_pass_math = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
percent_pass_math = (students_pass_math["math_score"].count()/total_students) *100

# Percentage of students with a passing reading score (70 or greater)
students_pass_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
percent_pass_reading = (students_pass_reading["reading_score"].count()/total_students)*100

# Percentage of students who passed math and reading (% Overall Passing)
overall_pass = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
overall_pass = (overall_pass["Student ID"].count()/total_students)*100

# Overall school results
school_summary = pd.DataFrame({"Total Schools": [total_school], "Total Students":[total_students],
                               "Total Budget": [total_budget], "Average Math Score": [avg_math],
                               "Average Reading Score":[avg_reading],"% Passing Math":[percent_pass_math],
                               "% Passing Reading":[percent_pass_reading],"% Overall Passing":[overall_pass]
                              })
school_summary
```

```python
# Format columns
school_summary["Total Students"]= school_summary["Total Students"].map("{:,}".format)
school_summary["Total Budget"]= school_summary["Total Budget"].map("${:,}".format)
school_summary["Average Math Score"]= school_summary["Average Math Score"].map("{:.2f}%".format)
school_summary["Average Reading Score"]= school_summary["Average Reading Score"].map("{:.2f}%".format)
school_summary["% Passing Math"]= school_summary["% Passing Math"].map("{:.2f}%".format)
school_summary["% Passing Reading"]= school_summary["% Passing Reading"].map("{:.2f}%".format)
school_summary["% Overall Passing"]= school_summary["% Overall Passing"].map("{:.2f}%".format)

school_summary
```

```python
school_data_complete.head()
```

```python
# Group data by schools in district
district_schools = school_data_complete.groupby("school_name")
district_schools
```

```python
# List school type
school_type= district_schools["type"].first()

# List total students per school
student_count = district_schools["Student ID"].count()

# List school student total
total_student_school = district_schools["student_name"].count()

# List school total budget
total_budget = school_data.groupby("school_name")["budget"].sum()

# List school budget per student
student_budget = total_budget/student_count

# List school average math score
avg_math_score = district_schools["math_score"].mean()

# List school average reading score
avg_read_score = district_schools["reading_score"].mean()

# Percent of students passing math per school
perc_pass_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()
perc_pass_math = (perc_pass_math["math_score"]/student_count) *100

# Percent of students passing reading per school
perc_pass_read = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()
perc_pass_read = (perc_pass_read["reading_score"]/student_count) *100

# Percent of students passing both reading and math per school
number_students_overall_passing = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name").count()
perc_overall_passing = (number_students_overall_passing["Student ID"]/student_count)*100
```

```python
# Create dataframe for math and reading calculations grouped by school
school_district_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_student_school,
    "Total School Budget": total_budget,
    "Per Student Budget": student_budget,
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_read_score,
    "% Passing Math": perc_pass_math,
    "% Passing Reading": perc_pass_read,
    "% Overall Passing": perc_overall_passing
})

school_district_df.index.name=None
school_district_df.head()
```

```python
# Create unformatted copy of dataframe and reset index 
school_district_data = school_district_df.reset_index(drop=True)
```

```python
# Format columns
school_district_df["Total Students"]= school_district_df["Total Students"].map("{:,}".format)
school_district_df["Per Student Budget"]= school_district_df["Per Student Budget"].map("${:.2f}".format)
school_district_df["Total School Budget"]= school_district_df["Total School Budget"].map("${:,}".format)
school_district_df["Average Math Score"]= school_district_df["Average Math Score"].map("{:.2f}%".format)
school_district_df["Average Reading Score"]= school_district_df["Average Reading Score"].map("{:.2f}%".format)
school_district_df["% Passing Math"]= school_district_df["% Passing Math"].map("{:.2f}%".format)
school_district_df["% Passing Reading"]= school_district_df["% Passing Reading"].map("{:.2f}%".format)
school_district_df["% Overall Passing"]= school_district_df["% Overall Passing"].map("{:.2f}%".format)
school_district_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_district_data.head()
```

```python
# Create bin to hold data for schools grouped by per student budget
spending_bin = [0, 585,630, 645, 676]
spending_groups = ["<$584", "$585-629", "$630-$644", "$645-675"]
```

```python
# Copy data from existing dataframe
spending_score = school_district_data

# Add new column for spending per student and set value to bins
spending_score["Spending Ranges (Per Student)"] = pd.cut(spending_score["Per Student Budget"], spending_bin, labels= spending_groups)

# Group by spending
spending_score = spending_score.groupby("Spending Ranges (Per Student)")

# Calculate scores based on school spending per student
math_score = spending_score["Average Math Score"].mean()
reading_score = spending_score["Average Reading Score"].mean()
percent_math= spending_score["% Passing Math"].mean()
percent_reading = spending_score["% Passing Reading"].mean()
overall_perc_passing = ((percent_math + percent_reading)/2)

# Create dataframe based on spending per student output values
spending_score = pd.DataFrame({
    "Average Math Score": math_score,
    "Average Reading Score": reading_score,
    "% Passing Math": percent_math,
    "% Passing Reading": percent_reading,
    "% Overall Passing": overall_perc_passing
})

spending_score
```

```python
# Format columns
spending_score["Average Math Score"]= spending_score["Average Math Score"].map("{:.2f}%".format)
spending_score["Average Reading Score"]= spending_score["Average Reading Score"].map("{:.2f}%".format)
spending_score["% Passing Math"]= spending_score["% Passing Math"].map("{:.2f}%".format)
spending_score["% Passing Reading"]= spending_score["% Passing Reading"].map("{:.2f}%".format)
spending_score["% Overall Passing"]= spending_score["% Overall Passing"].map("{:.2f}%".format)
spending_score
```
